```python
#declaramos los datos de nuestro interés

analfabetizmo = []
inasistenciaescolar = []
dependenciaeconom = []
fuenteaguamejorada = []


numerofilas = -1

#importar el archivo

with open('pobreza.csv', 'rb') as csvfile:
    archivo = csv.reader(csvfile, delimiter=' ', quotechar='"')
    for renglon in archivo:
       analfabetizmo.append(renglon[11])
       inasistenciaescolar.append(renglon[13])
       dependenciaeconom.append(renglon[21])
       fuenteaguamejorada.append(renglon[29])
       numerofilas=numerofilas+ 1
            

```

```python
#ahora llenemos nuestra matriz de datos

informacion= zeros((numerofilas,4))

for i in range(numerofilas):
    
    informacion[i][0] = analfabetizmo[1+i]
    informacion[i][1] = inasistenciaescolar[1+i]
    informacion[i][2] = dependenciaeconom[1+i]
    informacion[i][3] = fuenteaguamejorada[1+i]
    
```

```python
#vamos a calcular la amtriz de covarianza, pero primero calculemos los valores medios primero

shape(info)


numdatos = shape(informacion)[0]
dimensiones = shape(informacion)[1]



valoresmedios = zeros(dimensiones)

```

```python
#llenamos ahora la amtriz de covarianza


for i in range(dimensiones):
   valoresmedios[i] = mean(informacion[i,:])
    
   matrizcovarianza= zeros((dimensiones,dimensiones))

for m in range(dimensiones):
    for n in range(dimensiones):
        suma = 0
        for l in range(numdatos):
            suma=suma+ (informacion[l,m]-valoresmedios[m])*(informacion[l,n]-valoresmedios[n])
            matrizcovarianza[m,n] = suma/(numdatos-1)

```
