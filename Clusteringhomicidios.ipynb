```python

archivo = csv.reader(open('homicidios.1990.a.2013.csv','rU')) #obtener el archivo
numerofilas = 0 #numero filas del archivo
matrizdatos = [] 


for renglonenarchivo in archivo:
    
    if not "NA" in renglonenarchivo:
        if not "#DIV/0!"in renglonenarchivo:
            
            matrizdatos.append(renglonenarchivo)
            numerofilas=numerofilas+1
            
```

```python
informacion = np.zeros((numerofilas,24)) #nos interesan 24 columas de la matriz de datos


for i in range(1,numerofilas):
    j=0
    renglon = matrizdatos[i]
    
    while (4+(3*j) < size(renglon)):
        
            informacion[i,j]=renglon[5+(3*j)]
            j=j+1
           
            
#sólo tomamos los datos de 3 espacios en 3 espacios de la matriz de datos
```

```python


numdatos = shape(informacion)[0]
dimensiones = shape(informacion)[1]

valoresmedios= zeros(dimensiones)

for i in range(dimensiones):
    valoresmedios[i] = mean(informacion[:,i])


```

```python
matrizcovarianza = zeros((dimensiones,dimensiones)) 


for m in range(dimensiones):
    for n in range(dimensiones):
        suma = 0
        for l in range(numdatos):
            suma = suma+ (informacion[l,m]-valoresmedios[m])*(informacion[l,n]-valoresmedios[n])
            matrizcovarianza[m,n] = suma/(numdatos-1)

imshow(matrizcovarianza) #acá vemos la matriz simétrica de covarianza como era de esperar
```

```python

valorespropios = linalg.eig(matrizcovarianza)[0]


plot( valorespropios) #veamos si podemos reducir la dimensionalidad del asunto, ciertamente se ve que al menos 10 u 11 nos itneresan


#se discutió en clase el métodod e dar pesos a los valores apra saber cuáles despreciar

sumavalores= sum(valorespropios)
numvectoresbase=0
pesoimportancia = 0


for i in range(dimensiones):
    pesoimportancia = pesoimportancia + (valorespropios[i]/sumavalores)
    
    if (pesoimportancia>=0.89):
        numvectoresbase= i+1
        break

        
```

```python
#Ahora hagamos clustering para 2 centros con 5 iteraciones

#vector de 0 o 1 que me dice si se està en el prmer o segundo cluster

vectoridentificacion2 = zeros((numdatos))
vectoridentificacion2 = [randint(0,2) for i in range(numdatos)]


#iteraciones

for l in range(5):

    centros = zeros((2,dimensiones))
    tamano = zeros((2))


    for m in range(2):
        for n in range(numdatos):
            if m==vectoridentificacion2[n]:
                tamano[m] += 1            

    for m in range(numdatos):
        for n in range(2):
            if (n==vectoridentificacion2[m]):
                centros[n,:] += informacion[m,:]/tamano[n]
         #Calculo las distancias:

    distcentros = zeros((numdatos,2))
    distanciaminima = zeros((numdatos))


    for i in range(numdatos):
        for j in range(2):
            w = sum((informacion[i,:] - centros[j])**2)
            distcentros[i,j] = sqrt(w)
            
            #tomamos la distancia minima para elegir el nuevo centro
            distanciaminima[i] = argmin(distcentros[i,:])
        
    vectoridentificacion2 = distanciaminima
    
   
    #a contionuacion va la variacion de los centros para 5 iteraciones
```
