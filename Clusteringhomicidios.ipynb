```python

archivo = csv.reader(open('homicidios.1990.a.2013.csv','rU')) #obtener el archivo
numerofilas = 0 #numero filas del archivo
matrizdatos = [] 


for renglonenarchivo in archivo:
    
    if not "NA" in renglonenarchivo:
        if not "#DIV/0!"in renglonenarchivo:
            
            matrizdatos.append(renglonenarchivo)
            numerofilas=numerofilas+1
            
```

```python
informacion = np.zeros((numerofilas,24)) #nos interesan 24 columas de la matriz de datos


for i in range(1,numerofilas):
    
    j=0
    renglon = matrizdatos[i]
    
    while (size(renglon)>4+(3*j)):
        
            informacion[i,j]=renglon[5+(3*j)]
            j=j+1
           
            
#sólo tomamos los datos de 3 espacios en 3 espacios de la matriz de datos
```

```python

#vamos a sacar la matriz de covarianza para lo que primero se calcula los valores medios por facilidad
numdatos = shape(informacion)[0]
dimensiones = shape(informacion)[1]

valoresmedios= zeros(dimensiones)

for i in range(dimensiones):
    #còdigo que nos saca la media de un evector
    valoresmedios[i] = mean(informacion[:,i])


```

```python
#creacion de la matriz

matrizcovarianza = zeros((dimensiones,dimensiones)) 


for m in range(dimensiones):
    for n in range(dimensiones):
        suma = 0
        for l in range(numdatos):
            r=(informacion[l,m]-valoresmedios[m])*(informacion[l,n]-valoresmedios[n])
            suma = suma + r
            
            matrizcovarianza[m,n] = suma/(numdatos-1)

```

```python
#se discutió en clase el métodod e dar pesos a los valores apra saber cuáles despreciar

sumavalores= sum(valorespropios)
numvectoresbase=0
pesoimportancia = 0


for i in range(dimensiones):
    razon=(valorespropios[i]/sumavalores)
    pesoimportancia = pesoimportancia + razon
    
    if (pesoimportancia>=0.89):
        numvectoresbase= i+1
        break
```
